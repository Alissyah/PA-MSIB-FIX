# Data Exploration 
Data Exploration adalah sebuah kegiatan yang bertujuan untuk lebih memahami karakteristik dari data. Motivasi utama dilakukannya data exploration adalah untuk membantu kita untuk memilih tools yang tepat untuk melakukan preprocessing atau dalam melakukan analisis.

```python
def preprocess(df):
  for col in df.columns:
    arr = df[col].to_list()
    for i in range(len(arr)):
      low = arr[i].lower()
      w = low.replace(" ","") #Menghilangkan spasi pada dataset"
      df.loc[i,col] = w
  return df
```

```python
df = df_row.copy()
df = preprocess(df)
df
```

#Fungsi CountVectorizer, Cosine Similarity

CountVectorizer : CountVectorizer, berfungsi untuk menghitung frekuensi kata dalam dokumen. Count Vectorizer dapat mengubah fitur teks menjadi sebuah representasi vector. 

Penggunaan cosine similarity pada sistem ini adalah sebagai metode untuk membandingkan kesamaan atau similaritas perhitungan penggabungan inputan user : jenis lembaga, provinsi, umur, jenis kelamin, dan jenis kekerasan,  sistem rekomendasi berjalan mencari relevasi kemiripan berdasarkan inputan tersebut



```python
vect = CountVectorizer()#(stop_words='indonesian')
vect_matrix_features = vect.fit_transform(df['features'])
cosine_similarity_matrix_features = cosine_similarity(vect_matrix_features, vect_matrix_features)
```

```python
def index_from_title(df,title):
 return df[df['nama']==title].index.values[0]
```

#Modelling

```python
def recommendations_lembaga( nama,number_of_recommendations): 
  index = index_from_title(df,nama)
  similarity_scores = list(enumerate(cosine_similarity_matrix_features[index]))
  similarity_scores_sorted = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
  recommendations_indices = [t[0] for t in similarity_scores_sorted[1:(number_of_recommendations+1)]]

  print('Hasil rekomendasi lembaga lainnya adalah:')
  j = 1
  for i in recommendations_indices:
    print(j, df_row['nama'].iloc[i])
    j += 1
  #return df['nama'].iloc[recommendations_indices]
```

```python
def recommendations( jenis_lembaga, provinsi, jenis_kelamin, umur, jenis_kekerasan):
  provinsi = provinsi.replace(" ","")
  umur = umur.replace(" ","")
  input = (jenis_lembaga)+' '+(provinsi)+' '+(umur)+' '+(jenis_kelamin)+' '+(jenis_kekerasan)
  input = input.lower()
  index = df[df['features']==input].index.values 
  print('Hasil lembaga yang sesuai adalah:')
  for i in index: 
    print('Nama Lembaga: ',df_row.iloc[i,0])
    print('Jenis Lembaga: ',df_row.iloc[i,1])
    print('Profile Lembaga: ',df_row.iloc[i,2])
    print('Provinsi: ',df_row.iloc[i,3])
    print('Alamat: ',df_row.iloc[i,4])
    print('Kontak: ',df_row.iloc[i,5])
    print('Jenis Kelamin: ',df_row.iloc[i,6])
    print('Umur: ',df_row.iloc[i,7])
    print('Jenis Kekerasan: ',df_row.iloc[i,8])
    print('===================================================================')  
  return index
```
